```python
import tensorflow as tf
from imageio import imread,imsave
import matplotlib.pyplot as plt
import os
```

```python
label_colours = [[0, 0, 0], [153, 76, 0], [153, 153, 0]
                # 0 = road, 1 = sidewalk, 2 = building
                ,[76, 153, 0], [0, 153, 0], [0, 153, 76]
                # 3 = wall, 4 = fence, 5 = pole
                ,[0,153, 153], [0, 76, 153], [0, 0, 153]
                # 6 = traffic light, 7 = traffic sign, 8 = vegetation
                ,[76, 0, 153], [153, 0, 153], [153, 0, 76]
                # 9 = terrain, 10 = sky, 11 = person
                ,[255, 51, 51], [255, 153, 51], [255, 255, 51]
                # 12 = rider, 13 = car, 14 = truck
                ,[153, 255, 51], [51, 255, 51], [51, 255, 153]
                # 15 = bus, 16 = train, 17 = motocycle
                ,[51, 255, 255], [51,153,255],[51,51,255],[153,51,255],[255,52,255],[255,52,153], [192,192,192]]
                # 18 = bicycle
```

```python
def decode_labels(mask, img_shape, num_classes):

    color_table = label_colours

    color_mat = tf.constant(color_table, dtype=tf.float32)
    onehot_output = tf.one_hot(mask, depth=num_classes)
    onehot_output = tf.reshape(onehot_output, (-1, num_classes))
    pred = tf.matmul(onehot_output, color_mat)
    pred = tf.cast(tf.reshape(pred, (img_shape[0], img_shape[1], 3)),tf.uint8)
    
    return pred
```

```python
listof = ["104_42_c0001_segm_68.png","36_16_c0002_segm_34.png","40_05_c0028_segm_31.png","ung_104_22_c0009_segm_29.png","ung_133_10_c0001_segm_8.png",
          "104_52_c0002_segm_38.png","36_29_c0009_segm_97.png","40_06_c0007_segm_86.png","ung_104_46_c0001_segm_52.png","ung_133_25_c0001_segm_29.png",
          "138_10_c0001_segm_91.png","40_03_c0010_segm_60.png","40_06_c0031_segm_97.png","ung_104_49_c0013_segm_41.png","ung_138_53_c0002_segm_52.png",
          "138_51_c0002_segm_29.png","40_04_c0011_segm_3.png","40_08_c0002_segm_45.png","ung_133_03_c0001_segm_12.png","ung_60_04_c0006_segm_19.png",
          "36_04_c0001_segm_30.png","40_05_c0005_segm_95.png","ung_104_22_c0002_segm_25.png","ung_133_03_c0001_segm_8.png","ung_60_05_c0005_segm_71.png"]

list_gt = ["104_52_c0002_segm_38.png",
          "ung_133_25_c0001_segm_29.png",
          "36_16_c0002_segm_34.png",
          "36_04_c0001_segm_30.png"]

ap= "./final_final/out_2/"
ap_2 = "../../SURREAL/cmu/cluster_mod_test_heal/annotations/"

listo = []

for i in list_gt:
    listo.append(os.path.join(ap_2,i))

listy = []
for i in listof:
    listy.append(os.path.join(ap, i))
```

```python

im= tf.placeholder(tf.uint8, shape=(320,320))
ima = decode_labels(im, (320,320), 25)


with tf.Session() as sess:
    for i in listy:
        w = imread(i)
        name = i.split(".")
        namy = name[1]+ "_seg"
        f_name = "./"+".".join([namy, name[2]])
        f_im1= sess.run(ima,feed_dict={im:w})
    
        imsave(f_name,f_im1)
        
    
```
